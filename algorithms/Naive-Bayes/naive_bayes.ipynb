```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.naive_bayes import GaussianNB as sk_GaussianNB
from sklearn.naive_bayes import MultinomialNB as sk_MultinomialNB
from sklearn.naive_bayes import BernoulliNB as sk_BernoulliNB
```

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

gnb = GaussianNB()
gnb.fit(X_train, y_train)

preds = gnb.predict(X_test)
print(f"Test accuracy: {accuracy_score(y_test, preds):.3f}")

sk_gnb = sk_GaussianNB()
sk_gnb.fit(X_train, y_train)

preds = sk_gnb.predict(X_test)
print(f"Test accuracy (sklearn): {accuracy_score(y_test, preds):.3f}")
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer

categories = ["sci.space", "rec.sport.hockey", "talk.politics.guns"]
news = fetch_20newsgroups(subset="train", categories=categories,
                           remove=("headers", "footers", "quotes"))

vectorizer = CountVectorizer(max_features=2000, stop_words="english")
X_counts = vectorizer.fit_transform(news.data).toarray().astype(np.float64)
y_news = news.target

Xn_train, Xn_test, yn_train, yn_test = train_test_split(
    X_counts, y_news, test_size=0.25, random_state=42, stratify=y_news
)

mnb = MultinomialNB(alpha=1.0)
mnb.fit(Xn_train, yn_train)

preds = mnb.predict(Xn_test)
print(f"Test accuracy: {accuracy_score(yn_test, preds):.3f}")

sk_mnb = sk_MultinomialNB(alpha=1.0)
sk_mnb.fit(Xn_train, yn_train)

preds = sk_mnb.predict(Xn_test)
print(f"Test accuracy (sklearn): {accuracy_score(yn_test, preds):.3f}")
```

```python
from sklearn.datasets import load_breast_cancer

X_bc, y_bc = load_breast_cancer(return_X_y=True)
Xb_train, Xb_test, yb_train, yb_test = train_test_split(
    X_bc, y_bc, test_size=0.25, random_state=42, stratify=y_bc
)

medians = np.median(Xb_train, axis=0)
Xb_train_bin = (Xb_train > medians).astype(np.float64)
Xb_test_bin = (Xb_test > medians).astype(np.float64)

bnb = BernoulliNB(alpha=1.0)
bnb.fit(Xb_train_bin, yb_train)

preds = bnb.predict(Xb_test_bin)
print(f"Test accuracy: {accuracy_score(yb_test, preds):.3f}")

sk_bnb = sk_BernoulliNB(alpha=1.0)
sk_bnb.fit(Xb_train_bin, yb_train)

preds = sk_bnb.predict(Xb_test_bin)
print(f"Test accuracy (sklearn): {accuracy_score(yb_test, preds):.3f}")
```
